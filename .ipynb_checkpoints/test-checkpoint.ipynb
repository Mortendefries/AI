```python
# Checking for null values since it can be seen that there occur some within the price attribute
cab_df.isnull().sum()
```

```python
# Removing null values from the price attribute
cab_df.dropna(inplace=True)
cab_df.isnull().sum()
```

```python
# Removing id features (id and product_id) since we don't need them in our analysis
cab_df = cab_df[['distance', 'cab_type', 'time_stamp', 'destination', 'source', 'price',
                 'surge_multiplier', 'name']]
```

```python
cab_df.head(8)
```

```python
# Since there is both Uber and Lyft cab types, we check the allocation of the types
cab_df['cab_type'].value_counts()
```

```python
# To make an explorative analysis we divide the dataset up in one for Uber and one for Lyft
uber_df = cab_df[cab_df['cab_type'] =="Uber"]
lyft_df = cab_df[cab_df['cab_type'] =="Lyft"]
```

```python
uber_df["surge_multiplier"].value_counts()
```
